```python
import os
import shutil
import zipfile
```

```python
"""
# Este script verifica se as pastas listadas em 'items' est찾o presentes no diret처rio 'root'.
# Se encontradas, as pastas s찾o compactadas em arquivos ZIP e salvas no diret처rio 'new_root'.
"""
```

```python
root = r'C:\Users\Matheus\Pictures\pasta do model twin'
new_root = r'C:\Users\Matheus\Pictures\PRONTOS PARA SUBIR PROJETO C'
list_root = os.listdir(root)
print(list_root)
```

```python
items  = [
    "3PW_floor_3",
    "5PW_floor_5",
    "5PW_floor_3",
    "5PW_floor_2",
    "6PW_floor_1",
    "PW_floor_1_sticker",
    "PW_floor_4_sticker",
    "PW_floor_3_sticker",
    "PW_floor_2_sticker"
]

```

```python
for folder in list_root:
    if folder in items:
        folder_path = os.path.join(root, folder)
        if os.path.isdir(folder_path):  # Certifique-se de que 챕 uma pasta
            zip_file_path = os.path.join(new_root, f'{folder}.zip')
            # print(zip_file_path)
            
            with zipfile.ZipFile(zip_file_path, 'w') as zipf:
                for raiz, dirs, arquivos in os.walk(folder_path):
                    for arquivo in arquivos:
                        caminho_completo = os.path.join(raiz, arquivo)
                        print(caminho_completo,folder_path)
                        zipf.write(caminho_completo, os.path.relpath(caminho_completo, folder_path))
```
